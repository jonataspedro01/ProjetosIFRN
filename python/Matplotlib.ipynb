# Matplotlib

Matplotlib é uma biblioteca python para visualização de dados e geração de gráficos, dos mais básicos até os mais avançados;
Tem uma excelente integração com o jupyter notebook, permitindo a geração de gráficos no próprio jupyter, além de recursos como zoom;
Permite exportar gráficos para formatos de vetor, pdf, png, jpg entre várias outras;
É possível personalizar o gráfico e a geração pode ser feita através de funções ou mesmo de prog. Orientada a Objetos.

Site Oficial: https://matplotlib.org/

Instalação: pip install matplotlib

```python
# Importando a biblioteca principal
import matplotlib as mpl
```

```python
#Importando a biblioteca que facilita a geração dos gráficos
import matplotlib.pyplot as plt
```

```python
#Este comando serve para gerar os gráficos no próprio jupyter
%matplotlib inline
```

```python
plt.plot([1, 2, 3], [10, 5, 15])
plt.show()
```

```python
x = [1, 2, 3]
y = [7, 5, 7]
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3]
y = [7, 5, 7]
plt.plot(x, y)
plt.xlabel('Nome do eixo X')
plt.ylabel('Nome do eixo Y')
plt.title('Titulo do Gráfico')
plt.show()
```

```python
#Label - adiciona um rótulo ao gráfico (não aparece, é opcional)
#Legend() - exibe o rótulo/legenda do gráfico
x = [1, 2, 3]
y = [7, 5, 7]
plt.plot(x, y, label='Legenda do gráfico')
plt.legend()
plt.xlabel('Nome do eixo X')
plt.ylabel('Nome do eixo Y')
plt.title('Titulo do Gráfico')
plt.show()
```

```python
#grid() - adiciona as grades do gráfico
x = [1, 2, 3]
y = [7, 5, 7]
plt.plot(x, y, label='Legenda do gráfico')
plt.legend()
plt.xlabel('Nome do eixo X')
plt.ylabel('Nome do eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

Cores :
r - red
g - green
b - blue
y - yellow
c - cyan
k - preto
w - white

link das cores: https://matplotlib.org/3.1.0/gallery/color/named_colors.html

```python
#Adicionando cores
x = [1, 2, 3]
y = [7, 5, 7]
plt.plot(x, y, label='Legenda do gráfico', color='orange')
plt.legend()
plt.xlabel('Nome do eixo X')
plt.ylabel('Nome do eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

```python
#Inserindo 2 linhas no mesmo gráfico
x = [1, 2, 4]
y1 = [10, 4, 2]
y2 = [5, 10, 2]
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
#Inserindo legenda ao gráfico com as 2 linhas
x = [1, 2, 4]
y1 = [10, 4, 2]
y2 = [5, 10, 2]
plt.plot(x, y1, label='Resultado 1')
plt.plot(x, y2, label='Resultado 2')
plt.show()
```
